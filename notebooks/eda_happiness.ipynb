# ¿Qué hace felices a los países?
## Análisis exploratorio de los factores que explican la felicidad mundial en 2024

### Introducción
La felicidad de una población es un concepto complejo en el que influyen factores económicos, sociales, políticos y de salud. En los últimos años, diferentes informes internacionales han intentado medir este bienestar subjetivo para entender por qué algunos países muestran niveles de satisfacción vital más altos que otros.

En este proyecto se realizará un Análisis Exploratorio de Datos (EDA) a partir de un dataset internacional sobre felicidad, centrando el estudio en los datos correspondientes al año 2024. El objetivo no será solo describir qué países ocupan las primeras y últimas posiciones del ranking, sino también identificar qué variables parecen estar más relacionadas con un mayor nivel de felicidad.

A través de este análisis se buscará detectar patrones, comparar comportamientos entre países y extraer conclusiones útiles sobre los principales factores asociados al bienestar. De esta forma, el proyecto combinará una parte descriptiva con una interpretación orientada a negocio y toma de decisiones.

### Objetivo general
Analizar los datos de felicidad mundial de 2024 para identificar qué factores sociales, económicos y de calidad de vida están más relacionados con la puntuación de felicidad de los países, detectando patrones relevantes y diferencias entre los países con mejores y peores resultados.

### Hipótesis
1. Los países con mayor PIB per cápita tienden a presentar mayores niveles de felicidad.

2. El apoyo social es uno de los factores más influyentes en la felicidad de un país.

3. Una mayor esperanza de vida saludable está asociada con una mayor puntuación de felicidad.

4. Los países con mayor libertad para tomar decisiones vitales tienden a mostrar niveles más altos de felicidad.

5. Los países con mayores niveles de felicidad no destacan únicamente en una sola variable, sino en la combinación de varios factores favorables, como un mayor PIB per cápita, un fuerte apoyo social y una mejor esperanza de vida saludable.

## Limpieza ligera y preparación del dataset

En esta fase se revisarán posibles duplicados, se seleccionarán las variables más relevantes para el análisis, se renombrarán columnas largas y se construirá una versión simplificada del dataset orientada a la exploración y visualización.


```python
df_2024.duplicated().sum()
```

La revisión de duplicados permite comprobar si existen filas repetidas dentro del subconjunto de 2024. En caso de no encontrarse duplicados, se podrá continuar con el análisis sin necesidad de eliminar registros por este motivo.


```python
relevant_columns = [
    'Country name',
    'Rank',
    'Ladder score',
    'Explained by: Log GDP per capita',
    'Explained by: Social support',
    'Explained by: Healthy life expectancy',
    'Explained by: Freedom to make life choices',
    'Explained by: Generosity',
    'Explained by: Perceptions of corruption'
]

df_selected = df_2024[relevant_columns].copy()

df_selected.head()
```

En este punto se seleccionan únicamente las variables más útiles para responder a la pregunta principal del proyecto. Se conserva la puntuación de felicidad como variable objetivo, junto con las principales variables explicativas planteadas en las hipótesis.

```python
df_selected = df_selected.rename(columns={
    'Country name': 'country',
    'Rank': 'rank',
    'Ladder score': 'happiness_score',
    'Explained by: Log GDP per capita': 'gdp_per_capita',
    'Explained by: Social support': 'social_support',
    'Explained by: Healthy life expectancy': 'healthy_life_expectancy',
    'Explained by: Freedom to make life choices': 'freedom_of_choice',
    'Explained by: Generosity': 'generosity',
    'Explained by: Perceptions of corruption': 'perception_of_corruption'
})

df_selected.head()
```

El renombrado de columnas permite trabajar con nombres más breves, claros y consistentes. Esto facilita la lectura del notebook, simplifica la escritura del código y mejora la presentación de futuras visualizaciones.


```python
df_clean = df_selected.copy()

df_clean.info()
```

A partir de este momento, `df_clean` será el dataframe principal de trabajo para el resto del análisis.

## Análisis univariante de la felicidad

Una vez preparado el dataset, el siguiente paso consiste en explorar cómo se distribuye la puntuación de felicidad entre los países analizados en 2024.

Este bloque permitirá identificar:
- la distribución general de la variable principal,
- los países con mayores niveles de felicidad,
- los países con menores niveles de felicidad,
- y algunos patrones iniciales que servirán de base para análisis posteriores más profundos.


```python
df_clean['happiness_score'].describe()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df_clean['happiness_score'], bins=20, kde=True, color='teal')

plt.title('Distribution of Happiness Score in 2024', fontsize=14)
plt.xlabel('Happiness score')
plt.ylabel('Number of countries')
plt.show()
```

La distribución de la puntuación de felicidad permite observar cómo se reparten los niveles de bienestar entre los países analizados. Esta visualización ayuda a detectar si la mayoría de países se concentran en valores intermedios, si existen extremos muy marcados o si la distribución presenta cierta asimetría.


```python
top_10_happiest = df_clean.sort_values(by='happiness_score', ascending=False).head(10)

top_10_happiest[['country', 'happiness_score']]
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(
    data=top_10_happiest,
    x='happiness_score',
    y='country',
    hue='country',
    palette='viridis',
    legend=False
)

plt.title('Top 10 Happiest Countries in 2024', fontsize=14)
plt.xlabel('Happiness score')
plt.ylabel('Country')
plt.show()
```

Este ranking permite identificar qué países lideran la clasificación de felicidad en 2024. Además de mostrar las puntuaciones más altas, este resultado servirá más adelante para comparar si los países mejor posicionados comparten también patrones similares en variables como PIB, apoyo social o salud.


```python
saddest_10 = df_clean.sort_values(by='happiness_score', ascending=True).head(10)

saddest_10[['country', 'happiness_score']]
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(
    data=saddest_10,
    x='happiness_score',
    y='country',
    hue='country',
    palette='rocket',
    legend=False
)

plt.title('The 10 unhappiest countries of 2024', fontsize=14)
plt.xlabel('Happiness score')
plt.ylabel('Country')
plt.show()
```

La parte baja del ranking permite observar qué países presentan menores niveles de felicidad en 2024. Este grupo será especialmente útil para comparar sus características con las de los países mejor posicionados y detectar diferencias estructurales relevantes.


```python
happiest_saddest = pd.concat([top_10_happiest, saddest_10])

plt.figure(figsize=(14, 8))
sns.barplot(
    data=happiest_saddest.sort_values(by='happiness_score', ascending=False),
    x='happiness_score',
    y='country',
    hue='country',
    palette='coolwarm',
    legend=False
)

plt.title('Comparison Between the Happiest and Least Happy Countries in 2024', fontsize=15)
plt.xlabel('Happiness score')
plt.ylabel('Country')
plt.show()
```

La comparación conjunta entre los países más felices y menos felices permite apreciar de forma más clara la distancia existente entre ambos grupos. Esta diferencia sugiere que no se trata solo de pequeñas variaciones en la puntuación, sino de contextos nacionales potencialmente muy distintos, algo que se explorará en las siguientes fases del análisis.
