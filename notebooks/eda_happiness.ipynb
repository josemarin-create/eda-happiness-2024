# ¿Qué hace felices a los países?
## Análisis exploratorio de los factores que explican la felicidad mundial en 2024

### Introducción
La felicidad de una población es un concepto complejo en el que influyen factores económicos, sociales, políticos y de salud. En los últimos años, diferentes informes internacionales han intentado medir este bienestar subjetivo para entender por qué algunos países muestran niveles de satisfacción vital más altos que otros.

En este proyecto se realizará un Análisis Exploratorio de Datos (EDA) a partir de un dataset internacional sobre felicidad, centrando el estudio en los datos correspondientes al año 2024. El objetivo no será solo describir qué países ocupan las primeras y últimas posiciones del ranking, sino también identificar qué variables parecen estar más relacionadas con un mayor nivel de felicidad.

A través de este análisis se buscará detectar patrones, comparar comportamientos entre países y extraer conclusiones útiles sobre los principales factores asociados al bienestar. De esta forma, el proyecto combinará una parte descriptiva con una interpretación orientada a negocio y toma de decisiones.

### Objetivo general
Analizar los datos de felicidad mundial de 2024 para identificar qué factores sociales, económicos y de calidad de vida están más relacionados con la puntuación de felicidad de los países, detectando patrones relevantes y diferencias entre los países con mejores y peores resultados.

### Hipótesis
1. Los países con mayor PIB per cápita tienden a presentar mayores niveles de felicidad.

2. El apoyo social es uno de los factores más influyentes en la felicidad de un país.

3. Una mayor esperanza de vida saludable está asociada con una mayor puntuación de felicidad.

4. Los países con mayor libertad para tomar decisiones vitales tienden a mostrar niveles más altos de felicidad.

5. Los países con mayores niveles de felicidad no destacan únicamente en una sola variable, sino en la combinación de varios factores favorables, como un mayor PIB per cápita, un fuerte apoyo social y una mejor esperanza de vida saludable.
