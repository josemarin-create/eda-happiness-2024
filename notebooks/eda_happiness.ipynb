## Análisis univariante de la felicidad (Bloque descriptivo)

Una vez preparado el dataset, el siguiente paso consiste en explorar cómo se distribuye la puntuación de felicidad entre los países analizados en 2024.

- la distribución general de la variable principal,
- los países con mayores niveles de felicidad,
- los países con menores niveles de felicidad,

```python
df_clean['happiness_score'].describe()
```

La puntuación de felicidad en 2024 presenta una variabilidad considerable entre países. Aunque la media se sitúa en 5.58 puntos y la mediana en 5.87, la distancia entre el valor mínimo (1.364) y el máximo (7.736) refleja que existen diferencias muy marcadas en el nivel de bienestar percibido a escala internacional.

El hecho de que la media sea ligeramente inferior a la mediana sugiere una posible asimetría hacia valores bajos, probablemente influida por un grupo reducido de países con puntuaciones especialmente bajas.

La distribución de la felicidad muestra que la mayoría de países se concentra en niveles intermedios, mientras que los valores extremos son menos frecuentes. Sin embargo, la presencia de algunos países con puntuaciones especialmente bajas introduce una cierta asimetría en la distribución.

Esto sugiere que, aunque una parte importante del mundo se mueve en niveles medios de felicidad, existe un grupo reducido de países en situaciones claramente más desfavorables.

Los países con mayor puntuación de felicidad están encabezados por Finlandia, Dinamarca e Islandia.
```python
plt.figure(figsize=(12, 6))
sns.barplot(
    data=happiest_10,
    x='happiness_score',
    y='country',
    hue='country',
    palette='viridis',
    legend=False
)

plt.title('Top 10 Happiest Countries in 2024', fontsize=14)
plt.xlabel('Happiness score')
plt.ylabel('Country')
plt.show()
```

## Análisis bivariante: relación entre felicidad y factores explicativos

Una vez explorada la distribución general de la felicidad y los extremos del ranking, el siguiente paso consiste en analizar cómo se relaciona la puntuación de felicidad con algunas de las variables más relevantes del dataset.

En este bloque se estudiará la relación entre la felicidad y:

La matriz de correlación confirma que la felicidad está asociada positivamente con varias dimensiones del bienestar, especialmente apoyo social, PIB y salud. Esto refuerza la idea de que la felicidad responde a una combinación de factores y no a una única causa aislada.

## Análisis multivariante y comparación entre grupos

Después de estudiar la relación individual entre felicidad y varias variables explicativas, resulta útil dar un paso más y analizar cómo se combinan estos factores en los países con mejores y peores resultados.

En esta sección se compararán los países más felices y los menos felices para comprobar si los niveles altos de felicidad tienden a asociarse no con una sola variable, sino con la combinación de varios factores favorables al mismo tiempo.


```python
top_10_happiest = df_clean.sort_values(by='happiness_score', ascending=False).head(10).copy()
bottom_10_least_happy = df_clean.sort_values(by='happiness_score', ascending=True).head(10).copy()
```

```python
comparison_summary = pd.DataFrame({
    'Top 10 happiest': top_10_happiest[
        ['gdp_per_capita', 'social_support', 'healthy_life_expectancy', 'freedom_of_choice']
    ].mean(),
    'Bottom 10 least happy': bottom_10_least_happy[
        ['gdp_per_capita', 'social_support', 'healthy_life_expectancy', 'freedom_of_choice']
    ].mean()
})

comparison_summary
```

La comparación de medias permite resumir cómo se comportan las principales variables explicativas en los dos extremos del ranking de felicidad.

Si los países más felices presentan valores superiores de forma sistemática en varias dimensiones a la vez, esto reforzará la idea de que la felicidad está asociada a una combinación de factores favorables y no a un único elemento aislado.


```python
comparison_summary.plot(
    kind='bar',
    figsize=(12, 6),
    color=['mediumseagreen', 'indianred']
)

plt.title('Comparison of Key Factors: Top 10 vs Bottom 10 Countries', fontsize=14)
plt.xlabel('Variables')
plt.ylabel('Average value')
plt.xticks(rotation=45)
plt.legend(title='Group')
plt.show()
```

Este gráfico permite comparar visualmente las medias de las variables más relevantes entre los países con mayor y menor felicidad. La lectura conjunta de estas diferencias resulta especialmente útil para detectar si existe un patrón multivariante consistente en los países mejor posicionados.


```python
comparison_summary_reset = comparison_summary.reset_index()
comparison_summary_reset.columns = ['variable', 'Top 10 happiest', 'Bottom 10 least happy']

comparison_melted = comparison_summary_reset.melt(
    id_vars='variable',
    var_name='group',
    value_name='mean_value'
)

comparison_melted
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(
    data=comparison_melted,
    x='variable',
    y='mean_value',
    hue='group',
    marker='o',
    linewidth=2.5
)

plt.title('Multivariate Profile: Top 10 vs Bottom 10 Countries', fontsize=14)
plt.xlabel('Variables')
plt.ylabel('Average value')
plt.xticks(rotation=45)
plt.show()
```

La representación en forma de perfil facilita observar si ambos grupos siguen patrones diferentes de manera consistente. Si la línea del grupo de países más felices se mantiene por encima del grupo menos feliz en casi todas las variables, se refuerza la hipótesis de que la felicidad responde a una combinación de condiciones favorables.


## Comparación detallada de los países más y menos felices

Además de comparar medias grupales, también resulta útil observar cómo se distribuyen las variables clave dentro de cada país de los grupos extremos. Para ello se construye una visualización conjunta que permite detectar patrones comunes entre los países más felices y los menos felices.


```python
top_bottom_multivariate = pd.concat([
    top_10_happiest.assign(group='Top 10 happiest'),
    bottom_10_least_happy.assign(group='Bottom 10 least happy')
])

heatmap_data = top_bottom_multivariate.set_index('country')[
    ['gdp_per_capita', 'social_support', 'healthy_life_expectancy', 'freedom_of_choice']
]

plt.figure(figsize=(12, 10))
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt='.2f')

plt.title('Key Factors by Country: Top and Bottom Happiness Groups', fontsize=14)
plt.xlabel('Variables')
plt.ylabel('Country')
plt.show()
```

Este mapa de calor permite observar con más detalle cómo se comportan las variables clave dentro de cada país. Más allá de la comparación entre medias, esta visualización ayuda a comprobar si los países más felices comparten de manera consistente valores elevados en varias dimensiones al mismo tiempo.


## Validación de la hipótesis multivariante

Los resultados obtenidos en esta sección respaldan la hipótesis de que los países con mayores niveles de felicidad no destacan únicamente en una variable aislada, sino en la combinación de varios factores favorables.

La comparación entre los países más felices y los menos felices muestra que el grupo mejor posicionado presenta, de forma sistemática, valores medios más altos en PIB per cápita, apoyo social, esperanza de vida saludable y libertad para tomar decisiones vitales.

Esto sugiere que la felicidad no debe interpretarse como el resultado de una única causa, sino como la expresión de un contexto más amplio de bienestar económico, social y personal. En consecuencia, los datos apoyan una visión multivariante del fenómeno, donde diferentes dimensiones actúan conjuntamente.


Los países más felices no destacan por una fortaleza aislada, sino por un perfil global más sólido.
